```python
import tensorflow as tf
```

```python
t1 = tf.Variable([1, 2, 3])  # Variable은 mutable
t2 = tf.constant([1, 2, 3])  # 학습을 위해 만들어지는 constant는 immutable

# constant는 dataset에 대한 변수  ( 데이터셋을 바꿀 순 없으니까 )
# Variable은 model에 대한 변수 
# ( 모델 학습을 진행할 때마다 손실함수에 의해 새로운 가중치를 산정하므로 )
print(t1)
print(t2)

print(type(t1))
print(type(t2))
```

- list나 ndarray를 tensor로 변환 가능

```python
import numpy as np
# tf.constant
test_list = [ 1, 2, 3]
test_np = np.array([1,2,3])

t1 = tf.constant(test_list)
t2 = tf.constant(test_np)

print(t1)
print(t2)
print(type(t1))
print(type(t2))
```

```python
# tf.Variable

test_list = [ 1, 2, 3]
test_np = np.array([1,2,3])

t1 = tf.Variable(test_list)
t2 = tf.Variable(test_np)

print(t1)
print(t2)
print(type(t1))
print(type(t2))
```

```python
t1 = tf.constant(test_list)
t2 = tf.Variable(test_list)

t3 = tf.constant(t2)
t4 = tf.Variable(t1)

print(type(t3))
print(type(t4))

t5 = tf.convert_to_tensor(t3)

print(type(t5))
```

```python
test_list1 = [10, 20 ,30]
test_list2 = [1, 2, 3]

t1 = tf.constant(test_list1)
t2 = tf.Variable(test_list2)

t3 = t1 + t2
print(type(t3))


t4 = tf.Variable(test_list1)
t5 = tf.Variable(test_list2)
t6 = t4 + t5
print(type(t6))
```

- constant 와 Variable 간의 연산을 하면 constant Tensor 가 결과로 리턴
- Variable 과 Variable 간의 연산을 해도 constant Tensor 가 결과로 리턴

- TF에서는 정수보다는 실수로 만들어서 사용하는 것이 좋음.
- int 보다는 float 

```python
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev= 5)
t2 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)

t1 = tf.cast(t1, dtype=tf.int16)  # tf.cast : dtype을 변환하는 메서드
t2 = tf.cast(t2, dtype=tf.int16)
t3 = t1.numpy().astype('int16')  
# numpy() 함수로 ndarray로 변환한 후 astype해도 결과는 같음
print(t1.numpy())
print(t2.numpy())
print(t3)
```

```python
print( t1 + t2 )
```

```python
t1 = tf.cast(tf.random.normal(shape=(3, 1), mean=0, stddev=5), dtype=tf.int16)
t2 = tf.cast(tf.random.normal(shape=(1, 4), mean=0, stddev=5), dtype=tf.int16)

print(t1, '\n')
print(t2, '\n')
t3 = t1 + t2
print(t3)

# Broadcasting : 넘파이와 같이 한쪽의 요소가 모자라더라도 차원이 맞으면
# 알아서 확장하여 연산함
```

```python
tf.reduce_sum(t3).numpy()  # 옵션 지정을 안하면 모든 축의 값을 더 함
```

```python
print(tf.reduce_sum(t3, axis=1))
print(tf.reduce_sum(t3, axis=0))
# axis=1 축을 제거하는 옵션을 줘서 스칼라가 아닌 벡터로 계산 가능
# axis=1을 제거하면, 가로 행이 없어져서 (3, ) 차원의 벡터가 됨
# (3, 4) = (axis=0, axis=1) 로 이루어진 matrix에서 axis=1을 빼니까 
# axis=0으로만 이루어진 벡터가 된다는 것
```
