```python
df.rename(columns={'Store ID ':'가게 번호', 'Store_Area':'가게 면적', 'Items_Available':'판매품목 개수',
                   'Daily_Customer_Count':'일일 누적 고객', 'Store_Sales':'가게 매출'}, inplace=True)
df.set_index('가게 번호')
```

```python
df1 = df.drop('가게 번호', axis=1)
plt.rc('font', family='NanumGothic') # Windows 한글 font 설정
sns.pairplot(df1, diag_kind='hist')
plt.show()
```

```python
# vars=['컬럼명'] 패러미터로 원하는 컬럼끼리만 가능
sns.pairplot(df1, diag_kind='hist', vars=['일일 누적 고객', '가게 매출'])
plt.show()
```

```python
# 가게 매출을 기준으로 낮은 값부터 확인
df1.groupby(by=['가게 매출']).min()
```
