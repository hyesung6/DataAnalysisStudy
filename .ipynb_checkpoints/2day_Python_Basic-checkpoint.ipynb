## 연산 - 함수 적용

```python
# 날짜 타입으로 형 변환
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year+'-'+month+'-'+day
```

```python
d = 20210201
getDate(d)
```

```python
df['사용일자'].apply(getDate)
```

```python
df['사용일자1'] =  df['사용일자'].apply(getDate)
```

```python
# datetime64 형식으로 형변환 완료
df['사용일자1'].astype('datetime64')
```

```python
# 결측치 확인
df.isna()
```

## 변형

```python
df = pd.read_csv('CARD_SUBWAY_MONTH_202102.txt', index_col = False)
df.groupby(['노선명']).mean()
```

```python
df1 = pd.DataFrame({'이름': ['A', 'B', 'D'],
                    '성별':['남자', '여자', '남자']
                   })
df1
```

```python
df2 = pd.DataFrame({'이름': ['A', 'B', 'C'],
                    '키':[180, 160, 150],
                   })
df2
```

```python
# 교집합 join
pd.merge(df1, df2)
```

```python
# left 조인
# 좌측 df의 값은 누락없이 모두 join
pd.merge(df1, df2, how='left')
```

```python
# right 조인
# 우측 df 값은 누락 없이 모두 join
pd.merge(df1, df2, how='right')
```

```python
# outer 조인
# 양측의 요소를 모두 누락 없이 반영하는 합집합
pd.merge(df1, df2, how='outer')
```

### 연습문제

```python
df1 = pd.DataFrame({'이름': ['A', 'B', 'C', 'D', 'E'],
                    '키':[180, 160, 150, 170, 170],
                    '나이': [23, 46, 33, 65, 28],
                    '주소': ['서울', '부산', '대전', '원주', '제주']
                   })
df1
```

```python
df2 = pd.DataFrame({'이름': ['F', 'G', 'H', 'I', 'J'],
                    '키':[190, 180, 150, 160, 170],
                    '나이': [22, 18, 87, 45, 74],
                    '주소': ['전주', '서울', '수원', '부산', '인천']
                   })
df2
```

```python
# 컬럼명이 모두 동일한 데이터 형태인 경우
# 특정 기준값을 기준으로 합치는 것보다는
# 로우나 컬럼을 이어서 붙이는 형태로 합치는 것이 좋음
# 이어붙일 때 사용하는 메서드 concat()
# axis값을 안주면 기본적으로 row를 따라 이어붙인 데이터프레임 반환
pd.concat([df1, df2])
```

```python
# axis=1 값을 주면 컬럼을 따라 이어붙인 데이터프레임 반환
pd.concat([df1, df2], axis = 1)
```
