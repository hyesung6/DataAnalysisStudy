```python
df = data.get_data_yahoo('329180.KS', start='2021-01-01', end='2022-01-01')
df.reset_index(inplace=True)
```

```python
# Datetime split 1

df['Year'] = [s.year for s in df['Date']]
df['Month'] = [d.month for d in df['Date']]
df['Day'] = [d.day for d in df['Date']]
```

```python
# Datetime split 2
import datetime

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```
