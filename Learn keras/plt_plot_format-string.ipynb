matplotlib.pyplot.plot은 포맷 문자열을 지원한다.

color, marker, linestyle의 각 요소에 해당하는 포매팅 문자로 이루어진 문자열을 넘겨주면

자체적으로 각 파라미터를 지정하지 않아도 넣은 것과 같은 결과를 얻을 수 있음

fmt = 'color, marker, linestyle' 형식의 문자열

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_random_int(value:int):
    np.random.seed(value)
    data = [np.random.randint(10) for i in range(10)]
    return data
```

```python
X = range(1, 11)
data1 = get_random_int(1)
data2 = get_random_int(10)
data3 = get_random_int(100)
data4 = get_random_int(1000)
```

```python
plt.figure(figsize=(10, 7))
plt.title('Format string for plot : [color, marker, line]')
plt.plot(X, data1, 'c+-.', label='cyan, plus, dashed dot')
plt.plot(X, data2, 'm|--', label='magenta, vline, dashed')
plt.plot(X, data3, 'y,-', label='yellow, pixel, solid')
plt.plot(X, data4, 'k*:', label='black, star, dotted')
plt.xlabel('length')
plt.ylabel('Value')
plt.xticks(range(1, 11))
plt.legend()

plt.show()
```

- Color
  - b : blue
  - g : green
  - r : red
  - c : cyan
  - m : magenta
  - y : yellow
  - k : black
  - w : white

- Marker
  - . : point
  - , : pixel
  - o : circle
  - v : triangle down
  - ^ : triangle up
  - < : triangle left
  - \> : triangle right
  - 1 : triangle down2
  - 2 : triangle up2
  - 3 : triangle left2
  - 4 : triangle right2
  - s : square
  - p : pentagon
  - \* : star
  - h : hexagon1
  - H : hexagon2
  - \+ : plus
  - P : filled plus
  - x : x
  - D : diamond
  - d : thin diamond
  - | : vline
  - \- : hline

- line
  - \- : solid
  - \-- : dashed
  - \-. : dashdot
  - : : dotted
  - ' ' or '' : None
