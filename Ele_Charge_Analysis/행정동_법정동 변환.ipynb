```python
trade = pd.read_csv('당근마켓_전동휠체어_거래기록.csv', encoding='euc-kr')
```

```python
trade = trade.drop(['Unnamed: 0'], axis=1)
```

```python
used = trade
```

```python
# 중고거래 플랫폼 웹 크롤링 결과 df
## 법정동과 일치하지 않는 분할된 행정동만 조회
used[used['거래 장소'].str.contains('1|2|3')].drop_duplicates()
```

```python
import re
```

```python
addr = list(used['거래 장소'][used['거래 장소'].str.contains('1|2|3')].drop_duplicates())
```

```python
# 행정동을 법정동으로 변경하는 함수
def get_final_list(addr:list):
    addr_list = []
    for i in addr_:
        if '첨단' in i:      # 첨단1동은 법정동상 월계동이므로 추가적인 변환 필요 
            i = re.sub('첨단', '월계', i)     
        result = re.sub('1|2|3', '', i)
        addr_list.append(result)
    final = pd.DataFrame(addr_list, columns=['거래 장소'])
    final_list = final.drop_duplicates()
    return final_list
```

```python
get_final_list(addr)
```
