```python
# 인덱스 컬럼을 0부터 순차적으로 증가하는 정수로 변경하세요.
# 새로운 인덱스를 추가하는 방법
df.reset_index(inplace=True)
df
```

```python
df.rename(columns = {'Date':'날짜', 'High':'고가', 'Low':'저가'}, inplace=True)
```

```python
df.drop(['날짜'], axis=1)
```

```python
df['diff'] = df['Close'] - df['Open']
df
```

```python
df['Year'] = [d.year for d in df['날짜']]
df['Month'] = [d.month for d in df['날짜']]
df['Day'] = [d.day for d in df['날짜']]
df
```

```python
format = lambda x: '%.2f' % 2
df.applymap(format)
```

```python
df['거래량'] = df['Volume'].apply(lambda x: "{:,}".format(x))
```

```python
volume_mean = df['Volume'].groupby(df['Month']).mean()
volume_mean
```

```python
df.isna().sum()
```

```python
df.dropna()   # 결측치가 있을 경우 예시
df.fillna({'High':90000, 'Low':50000})  # 결측치가 있을 경우 예시
```

```python
dec_df = df[df['Month'] == 12]
nov_df = df[df['Month'] == 11]

nov_dec_df = pd.concat([nov_df, dec_df])
nov_dec_df
```
