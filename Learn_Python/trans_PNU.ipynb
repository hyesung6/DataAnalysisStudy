# 울진_죽변_화성

```python
result = []

죽변_화성.sort_index(inplace=True)
죽변_화성['지번'] = 죽변_화성['지번'].astype('string')

for i in 죽변_화성['지번']:
    if i[0] == '산':
        산여부 = '2'
    else:
        산여부 = '1'
    
    try:
        a = i.split('-')
        if a[0][0] == '산':
            a[0] = a[0].replace('산', '')
        else:
            pass
    
        if a[0] == None:
            a[0] == '0000'
        elif len(a[0]) == 1:
                a[0] = '000'+str(a[0])
        elif len(a[0]) == 2:
                a[0] = '00'+str(a[0])
        elif len(a[0]) == 3:
                a[0] = '0'+str(a[0])

        if a[1] == None:
            a[1] == '0000'
        elif len(a[1]) == 1:
            a[1] = '000'+str(a[1])
        elif len(a[1]) == 2:
            a[1] = '00'+str(a[1])
        elif len(a[1]) == 3:
            a[1] = '0'+str(a[1])
        result.append(산여부+a[0]+a[1])
        
    except:
        b = i
        if b[0] == '산':
            b = b.replace('산', '')
        else :
            pass
        
        if b == None:
            b == '0000'
        elif len(b) == 1:
                b = '000'+str(b)
        elif len(b) == 2:
                b = '00'+str(b)
        elif len(b) == 3:
                b = '0'+str(b)
        result.append(산여부+b+'0000')
```

```python
PNU = pd.DataFrame(result, columns=['PNU'])

죽변_화성['PNU'] = PNU

죽변_화성['시군구코드'] = str(dong_code[(dong_code['시군구명'] == '울진군') & (dong_code['읍면동명'] == '죽변면') & (dong_code['동리명'] == '화성리')].법정동코드.values)[1:11]

죽변_화성['PNU_Final'] = 죽변_화성['시군구코드'] + 죽변_화성['PNU']

죽변_화성.drop(['PNU', '시군구코드'], inplace=True, axis=1)

죽변_화성.rename(columns={'PNU_Final':'PNU'}, inplace=True)

죽변_화성['PNU'] = 죽변_화성['PNU'].astype('object')

죽변_화성.to_excel('죽변_화성_PNU.xlsx')
```
